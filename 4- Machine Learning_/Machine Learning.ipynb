```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Presentación 
games =  pd.read_csv('games.csv')
print(games.shape)
games.head()

#Config
pd.set_option('display.max_rows', 10000)



# Transfomraciónes
games["log_Critic_count"] = np.log(games["Critic_Count"])
games["log_Critic_Score"] = np.log(games["Critic_Score"])




# -- NaN -

## Reemplazo con Media ##
reemplazar_con_medias = list(games.iloc[:, -2:].columns.values)
reemplazar_con_medias.append("User_Count")


for i in  reemplazar_con_medias:
    games[f"{i}_is_bad"] = np.where(games[f"{i}"].isna(),1 ,0)
    games[f"{i}"].fillna(value=games[f"{i}"].mean(), inplace=True)
  

## Elimino columnas##
games.drop(columns=["Critic_Count", "Critic_Score", "Rating", "User_Score", "Developer", "Name"], axis=1, inplace=True)

## Elimno Filas ## 
games.dropna( inplace=True)



#Crear Columnas Nuevas
count = games.Publisher.value_counts()
small = count[count<=50]
medium  =  count[(50<count) & (count<=132)]
big =  count[(132<count) & (count<=300)]


games["Small_Publisher"] = np.where(games["Publisher"].isin(small.index.values),1, 0 )
games["Medium_Publisher"] = np.where(games["Publisher"].isin(medium.index.values),1, 0 )
games["Big_Publisher"] = np.where(games["Publisher"].isin(big.index.values),1, 0 )





# Elimino Variables Object 
games.drop("Publisher", axis=1 , inplace=True)

games.to_csv("games_clean.csv", index=False)
```
