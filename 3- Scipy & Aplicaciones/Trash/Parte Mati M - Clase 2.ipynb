#### Vamos a comenzar nuestro trabajo generando valores aleatorios con una distribucion determinada

### Distribucion Binomial

__construimos una variable llamada "a" con distribucion Binomial__

$$ a∼Bi(n=10,p=0.2) $$

__Recordemos: la distribucion Binomial tiene una funcion conocida de este estilo:__

$$ P(a) = \binom{x}{a}p^a(1-p)^{x-a} $$

_de la cual podemos inferir muchos resultados, comencemos..._

```python
a = stats.binom(10, 0.5)
sample = a.rvs(size=100)
sample
```

### Graficamos la variable y comenzamos a ver las funciones que nos ofrece este modulo:

__*repasemoslas:*__

* pmf(k, n, p, loc=0)
    Probability mass function.

* cdf(k, n, p, loc=0)
    Cumulative distribution function.

* sf(k, n, p, loc=0)
    Survival function  (also defined as ``1 - cdf``.

* ppf(q, n, p, loc=0)
    Percent point function (inverse of ``cdf`` --- percentiles).

* stats(n, p, loc=0, moments='mv')
    Mean('m'), variance('v'), skew('s'), and/or kurtosis('k').

* median(n, p, loc=0)
    Median of the distribution.

* mean(n, p, loc=0)
    Mean of the distribution.

* var(n, p, loc=0)
    Variance of the distribution.

* std(n, p, loc=0)
    Standard deviation of the distribution.

* interval(alpha, n, p, loc=0)
    Endpoints of the range that contains alpha percent of the distribution

### (CDF) Funcion de Distribucion Acumulada

```python
cdf_binomial = [a.cdf(i) for i in np.arange(11)]
df = pd.Series(cdf_binomial)
print(df.plot.bar())
print(a.cdf(1))
print(a.cdf(5))
print(a.cdf(9))
```

### (PMF) Funcion de Probabilidad Puntual

```python
pmf_binomial = [a.pmf(i) for i in np.arange(11)]
df = pd.Series(pmf_binomial)
print(df.plot.bar())
print(a.pmf(1))
print(a.pmf(5))
print(a.pmf(9))
```

### (SF) Funcion de Supervivencia

```python
sf_binomial = [a.sf(i) for i in np.arange(11)]
df = pd.Series(sf_binomial)
print(df.plot.bar())
print(a.sf(1))
print(a.sf(5))
print(a.sf(9))
```

```python
a.sf(5) + a.cdf(5)
```

### (MVSK) Los 4 indicadores descriptivos: Media, Varianza, Simetria y Curtosis

```python
a.stats('mvsk')
```

### Mediana y Momentos

```python
a.median()
```

```python
print(a.moment(1))
print(a.moment(2))
print(a.moment(1)**2)
print(a.moment(2)-a.moment(1)**2)
```

```python
a.interval(0.5)
```
